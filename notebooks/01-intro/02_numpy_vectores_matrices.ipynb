### Conclusiones Preliminares

* La diferencia entre una `lista` Python y un `array` de NumPy, es significativa en términos de funcionalidad, rendimiento y uso en la práctica.

**Tipo de dato**

* Una lista en Python es una colección genérica que puede contener elementos de **diferentes tipos**. Por ejemplo: strings, int, float, etc.
* Los elementos de una lista son **objetos** en Python. Esto implica un mayor uso de memoria y menor eficiencia en operaciones numéricas.

* Por su parte, un array de NumPy es una estructura de datos homogénea, es decir, todos sus elementos deben ser del mismo tipo. 
* Los elementos de un array de NumPy se almacenan en un bloque contiguo de memoria, lo que permite operaciones más eficientes.

**Rendimiento**

* Las operaciones en lista de Python son más lentas porque están diseñadas para ser flexibles y manejar cualquier tipo de dato.
* Por ejemplo, sumar dos listas requiere un bucle explícito o el uso de funciones como `map` o `list comprehension`.


* Por su parte, NumPy está optimizado para operaciones numéricas. Las operaciones en arrays de NumPy se ejcutan en tiempo constante y están altamente optimizadas gracias a su implementación en C.
* Por ejemplo, al sumar dos arrays de NumPy es una operación vectorizada que se ejecuta mucho más rápido que en listas de Python.


**Operaciones Vectorizadas**

* La lista de Python no admite operaciones vectorizadas de manera nativa. Por ejemplo, para sumar dos listas elemento por elemento, necesitas usar un bucle o una comprensión de listas.

* Por su parte, un array de NumPy puede realizar operaciones en todo el array sin necesidad de crear bucles explícitos.

```python
# Operaciones vectorizadas en listas

list1 = [1, 2, 3]

list2 = [4, 5, 6]

result = [a + b for a, b in zip(list1, list2)]

print(result) 
```

```python
# Operaciones vectorizadas en numpy


array1 = np.array([1, 2, 3])

array2 = np.array([4, 5, 6])

result = array1 + array2

print(result)  
```

## Números aleatorios

Numpy tiene diferentes formas de crear arreglos de números aleatorios. El módulo para realizar esto se llama Random.

#### rand

Crea un arreglo de la forma dada y la rellena con muestras aleatorias de una distribución uniforme sobre [0, 1].

```python
# creación de un arreglo de 2 elementos de 1 dimensión

np.random.rand(2)
```

#### randint

Entrega números enteros aleatorios.

```python
np.random.randint(1, 100)
```

```python
np.random.randint(1, 100, (10, 2))
```

## Concatenación

La función `np.concatenate` se usa para unir dos o más arrays a lo largo de un eje especificado. Si no se especifica el eje, por defecto se concatena a lo largo del **primer eje** (axis=0), que en un array 2D corresponde a las filas.

```python
# En una dimension
x = np.array([1, 2, 3])  # Vector de valores
y = np.array([3, 2, 1])  # Vector de valores
np.concatenate([x, y])  # Concatenación
```

```python
# En dos dimensiones de forma vertical
grid = np.array([[1, 2, 3], [4, 5, 6]])
print(grid)
print("----")
print(np.concatenate([grid, grid], axis=0))  # Concatenación
print("----")
print(np.concatenate([grid, grid], axis=1))
```
