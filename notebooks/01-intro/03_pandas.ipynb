# Introducción a Pandas

Pandas es una poderosa herramienta de manipulación y análisis de datos en Python, desarrollada por [Wes McKinney](https://wesmckinney.com/). Inicialmente construida sobre Numpy y ahora compatible con tecnologías como Arrow, Pandas facilita la limpieza y transformación de datos en bruto, convirtiéndolos en formatos estructurados y aptos para el análisis.

Originalmente diseñada para gestionar datos financieros y ofrecer una alternativa a las hojas de cálculo tradicionales (como Microsoft Excel), se ha consolidado como una librería esencial para el procesamiento de datos y la estadística en Python.

## Tipos de Datos en Pandas

Pandas permite trabajar con distintos tipos de datos, entre los que destacan:

- **Datos tabulares:** Tablas con columnas heterogéneas y etiquetas tanto en filas como en columnas.
- **Series temporales:** Estructuras especializadas para el manejo de datos cronológicos.

## Funcionalidades Clave

Entre las principales funcionalidades de Pandas se incluyen:

- **Lectura y escritura de datos:** Soporta múltiples formatos, como CSV, JSON, Excel, bases de datos SQL, Parquet y HDF5, entre otros.
- **Selección y filtrado:** Permite extraer subconjuntos de datos basados en la posición, valores o etiquetas de forma sencilla.
- **Fusión y unión de datos:** Facilita la integración y combinación de diferentes conjuntos de datos.
- **Transformación de datos:** Ofrece herramientas para aplicar funciones tanto de forma global como a través de ventanas móviles.
- **Manipulación de series temporales:** Proporciona métodos especializados para el tratamiento y análisis de datos basados en el tiempo.
- **Visualización:** Incorpora funcionalidades para generar gráficos y explorar visualmente los datos.

## Estructuras de Datos Fundamentales

Pandas se basa en dos objetos principales:

- **Series:** Estructuras unidimensionales similares a listas.
- **DataFrame:** Estructuras bidimensionales que representan tablas de datos.

En resumen, Pandas ofrece las estructuras de datos y las funciones necesarias para transformar y analizar información de manera eficiente, consolidándose como una herramienta indispensable en el ecosistema Python para la ciencia de datos y la estadística.

# Series en Pandas

Una **Series** es uno de los tipos de datos fundamentales en Pandas. Es similar a un arreglo de NumPy, pero añade la ventaja de tener etiquetas en los ejes, lo que permite indexarla por etiquetas en lugar de usar únicamente posiciones numéricas. Además, una Series no se limita a contener datos numéricos; puede almacenar cualquier tipo de objeto de Python.

## Creación de una Series

Puedes crear una Series a partir de diferentes estructuras de datos (listas, arreglos de NumPy o diccionarios) utilizando el método `pd.Series`.

### Ejemplo: Preparación de Datos

```python
# Definición de datos

labels = ["a", "b", "c"]          # Lista de etiquetas
my_list = [10, 20, 30]            # Lista de valores
arr = np.array([10, 20, 30])      # Arreglo de NumPy
d = {"a": 10, "b": 20, "c": 30}    # Diccionario
```

```python
# Tamaño en memoria

print(f"Lista 'labels': {sys.getsizeof(labels)} bytes")
print(f"Lista 'my_list': {sys.getsizeof(my_list)} bytes")
print(f"Array NumPy: {arr.nbytes} bytes (solo datos) + {sys.getsizeof(arr)} bytes (objeto completo)")
print(f"Diccionario: {sys.getsizeof(d)} bytes")
```

### Creación desde una lista

#### 1. Sin especificar etiquetas:

Al crear una Series a partir de una lista, se asignan automáticamente índices numéricos.

```python
my_list
```

```python
pd.Series(data=my_list)
```

#### 2. Con etiquetas personalizadas:

Puedes especificar etiquetas usando el parámetro `index`.

```python
labels
```

```python
pd.Series(data=my_list, index=labels)
```

#### 3. Sintaxis abreviada

No es obligatorio usar el nombre del parámetro `data`.

```python
pd.Series(my_list, labels)
```

#### 4. Creación desde un arreglo de NumPy

```python
# con etiquetas

pd.Series(arr, labels)
```

```python
# sin etiquetas

pd.Series(arr)
```

#### 5. Creación desde un Diccionario

Al crear una Serie a partir de un diccionario, las claves se usan automáticamente como etiquetas.

```python
d
```

```python
pd.Series(d)
```

#### 6. Serie con datos no numéricos

Una serie puede contener cualquier tipo de dato. Por ejemplo, crear una Series a partir de una lista de letras.

```python
pd.Series(data=labels)
```

## Indexación y operaciones en una `Series`

El índice es fundamental en una Series, ya que permite acceder a los datos de manera rápida y eficiente, similar al funcionamiento de un diccionario.

```python
# Creación de series

ser1 = pd.Series([1, 2, 3, 4], index=["USA", "Alemania", "Italia", "Japón"])
print(ser1)
print("-------------------------------------------------------------------------")
ser2 = pd.Series([1, 2, 5, 4], index=["USA", "Alemania", "Colombia", "Japón"])
print(ser2)
```

```python
# Acceso a elementos

print(ser1["USA"])
```

```python
# Las operaciones también se realizan según el índice

ser1 + ser2
```

**Nótese que las etiquetas que no se encuentran en ambas resultan en `NaN`.**

## Groupby (Agrupación por filas)

El método groupby le permite agrupar filas de datos y llamar a funciones agregadas.

```python
# Crear dataframe desde un diccionario
data = {
    "Company": ["GOOG", "GOOG", "MSFT", "MSFT", "FB", "FB", "GOOG", "MSFT", "FB"],
    "Person": [
        "Sam",
        "Charlie",
        "Amy",
        "Vanessa",
        "Carl",
        "Sarah",
        "John",
        "Randy",
        "David",
    ],
    "Sales": [200, 120, 340, 124, 243, 350, 275, 400, 180],
}

# Conversión del diccionario a dataframe
df = pd.DataFrame(data)
df
```

Se puede usar el método `.groupby()` para agrupar filas en función de un nombre de columna. Por ejemplo, vamos a agruparnos a partir de la compañía. Esto creará un objeto `DataFrameGroupBy`:

```python
# Agrupar por compañía

df.groupby("Company")
```

```python
# utilizar los métodos agregados del objeto

df.groupby("Company").mean(numeric_only=True)
```

```python
# Agrupar por compañía y calcular la desviación estándar

df.groupby("Company").std(numeric_only=True)
```

```python
# Agrupar por compañía y calcular el mínimo

df.groupby("Company").min(numeric_only=True)
```

```python
df
```

```python
# Agrupar por compañía y sumar los elementos que hay excluyendo los NaN

df.groupby("Company").count()
```

Una de las funciones más poderosas para descripción estadística de un dataframe es `describe()`

```python
df.groupby("Company").describe()
```

## Pivot Table

La funcionalidad "pivot_table" es muy utilizada y popular en las hojas de cálculo. Esta funcionalidad permite agrupar, ordenar, calcular datos. Para más información sobre pivot table hacer clic [aquí](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html).

```python
import seaborn as sns
```

```python
# Se carga el dataset de titanic

titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# Si se hiciera una pivot_table a mano para obtener el promedio de personas que sobrevivieron por género:

titanic.groupby("sex")[["survived"]].mean()
```

```python
# Para obtener el promedio de cuántos sobrevivieron por género divididos por clase:

titanic.groupby(["sex", "class"], observed=True)[["survived"]].mean().unstack()
```

* Con el parámetro `observed=True` solo se incluyen las categorías que están presentes en los datos. Evita incluir filas o columnas vacías en el resultado final.

* Con el método `unstack()` se transforma (o "desapila") uno de los niveles del índice en columnas. Cuando se utiliza en el resultado de un groupby() que genera un MultiIndex (por ejemplo, un índice jerárquico basado en "sex" y "class"), `unstack()` convierte uno de esos niveles en columnas para facilitar la visualización y el análisis.

### Usando Pivot Tables

```python
titanic.pivot_table("survived", index="sex", columns="class", observed=True)
```

## Concatenar y fusionar

Concatenar y fusionar son las dos formas principales de combinar DataFrames.

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)
df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
```

### Concatenación

* La concatenación básicamente combina DataFrames. Se debe tener en cuenta que las dimensiones deben coincidir a lo largo del eje con el que se está concatenando.

* La concatenación se hace con dataframes de diferentes índices.

* Puede usar `.concat()` y pasar una lsita de DataFrames para concatenar juntos.

```python
pd.concat([df1, df2, df3], axis = "index")
```

También se puede concatenar el DataFrame horizontalmente.
**Nota:** como no coinciden los index, observa lo que ocurre.

```python
pd.concat([df1, df2, df3], axis="columns")
```

### Fusión (merging)

La fusión `merge()` le permite fusionar DataFrames juntos utilizando una lógica similar a la combinación de Tablas SQL. Por ejemplo:

```python
# DataFrames de ejemplo para merging
left = pd.DataFrame(
    {
        "Producto": ["Arepas", "Banano", "Cafe"],
        "Tienda": [1, 2, 1],
        "Ventas": [100, 200, 50],
    }
)

right = pd.DataFrame(
    {
        "Producto": ["Arepas", "Banano", "Cafe"],
        "Tienda": [1, 2, 1],
        "Inventario": [50, 100, 75],
    }
)
```

```python
left
```

```python
right
```

```python
# how="inner" utiliza la intersección de las claves de ambos marcos,
# similar a una combinación interna de SQL.

pd.merge(left, right, how="inner", on="Producto")
```

Un ejemplo más complicado:

* **Natural join:** para mantener solo las filas que coinciden con los marcos de datos, especifique el argumento **how = ‘inner’**.
* **Full outer join:** para mantener todas las filas de ambos dataframe, especifique **how = ‘OUTER’**.
* **Left outer join:** para incluir todas las filas de su dataframe x y solo aquellas de y que coincidan, especifique **how=‘left’**.
* **Right outer join:** para incluir todas las filas de su dataframe y y solo aquellas de x que coincidan, especifique **how=‘right’**.


Imagen para entender los tipos de merge [aquí](https://miro.medium.com/v2/resize:fit:752/0*ZGDgg-0SznYlwiij.png)

```python
left = pd.DataFrame(
    {
        "Producto": ["Arepas", "Leche", "Cafe"],
        "Tienda": [1, 2, 1],
        "Ventas": [100, 200, 50],
    }
)

right = pd.DataFrame(
    {
        "Producto": ["Arepas", "Banano", "Cafe"],
        "Tienda": [1, 2, 1],
        "Inventario": [50, 100, 75],
    }
)
```

```python
left
```

```python
right
```

#### Claves comunes entre tablas

```python
# Fusionando comparando las mismas claves que tengan comunes

pd.merge(left, right, on=["Producto"])
```

#### Fusionando totalmente las dos tablas con las claves

```python
pd.merge(left, right, how="outer", on=["Producto"])
```

#### Fusionando usando las claves de la tabla right

```python
print("Tabla left")
print(left)
print("-------------------------------------------------------------------------")

print("Tabla right")
print(right)
```

```python
pd.merge(left, right, how="right", on=["Producto"])
```

#### Fusionando usando las claves de la tabla left

```python
pd.merge(left, right, how="left", on=["Producto"])
```

# Correción de tipos de datos (casting)

Cuando se trabaja con datos, es posible que se encuentre con un DataFrame donde el tipo de datos de una columna no es la correcta. Los tipos de datos son principalmente: numéricos, categóricos nominales, categóricos ordinales, booleanos, fechas y tiempos, texto.

Se debe convertir los tipos de datos para que sean apropiados para el tipo de datos que representan. Por ejemplo, es posible que desee convertir una columna de strings que representan números en números enteros o flotantes. O puede tener una columna de números que representan fechas, pero que pandas no reconoce como fechas.

**Nota: al convertir los datos a su forma correcta, normalmente hace que los DataFrames sean mucho más pequeños en la memoria, lo que los hace más eficientes de usar.**

```python
import seaborn as sns  # importar la librería seaborn

titanic = sns.load_dataset("titanic")
titanic.info()
```

```python
titanic.sample(5)
```

## Casting variables categóricas nominales

```python
# convertir la columna sex, embarked, who, embark_town en categóricas
cols_categoricas_nom = ["sex", "embarked", "who", "embark_town"]

# ver los datos únicos de las columnas_categoricas_nom
for columna in cols_categoricas_nom:
    print(f"Valores únicos en {columna}: {titanic[columna].unique()}")
```

```python
titanic[cols_categoricas_nom] = titanic[cols_categoricas_nom].astype("category")

# ver los datos únicos de las columnas_categoricas_nom
for columna in cols_categoricas_nom:
    print(f"Valores únicos en {columna}: {titanic[columna].unique()}")
```

## Casting variables categóricas ordinales

```python
cols_categoricas_ord = ["pclass", "deck"]

for columna in cols_categoricas_ord:
    print(f"Valores únicos en {columna}: {titanic[columna].unique()}")
```

```python
titanic["pclass"] = pd.Categorical(
    titanic["pclass"], categories=[3, 2, 1], ordered=True
)

titanic["deck"] = pd.Categorical(
    titanic["deck"], categories=list("ABCDEFG"), ordered=True
)
```

```python
for columna in cols_categoricas_ord:
    print(f"Valores únicos en {columna}: {titanic[columna].unique()}")
```

## Casting variables booleanas

```python
columnas_bool = ["alive", "alone", "survived"]

titanic[columnas_bool] = titanic[columnas_bool].astype("bool")
```

# Datos faltantes

```python
# data frame con algunos datos faltantes
# NaN = Not a Number
df = pd.DataFrame({"A": [1, 2, np.nan], "B": [5, np.nan, np.nan], "C": [1, 2, 3]})
df
```

```python
# Detectar si faltan datos

df.isna()
```

```python
# Verificar si hay datos faltantes por columna

df.isna().sum()
```

```python
# Eliminar datos faltantes

df.dropna(axis="columns")
```

```python
# eliminar las filas que tengas 2 o mas valores NaN
df.dropna(thresh=2)
```

```python
# Llenar los datos faltantes con el dato que nos interese
df.fillna(value="Llenar Valores")  # llenar los espacios con un string
# puede ser una palabra, numero , etc
```

```python
# Llenar los datos faltantes con el dato que nos interese
df.fillna(value=99)  # llenar los espacios con un numero
```

```python
# Llenar los datos faltantes con el promedio de esa columna
df["A"].fillna(value=df["A"].mean())
```

# Tablas de contingencia

```python
# Crear Datos
data = pd.DataFrame(
    {
        "gender": [
            "male",
            "male",
            "female",
            "female",
            "male",
            "female",
            "male",
            "female",
        ],
        "education_level": [
            "high school",
            "college",
            "college",
            "graduate",
            "high school",
            "graduate",
            "college",
            "graduate",
        ],
        "city": [
            "Boston",
            "Boston",
            "Miami",
            "Boston",
            "Miami",
            "Boston",
            "Miami",
            "Miami",
        ],
        "score": [75, 82, 88, 95, 69, 92, 78, 85],
    }
)

data
```

Una **tabla de contingencia** (o tabla cruzada) es una herramienta que resume la relación entre dos o más variables categóricas, mostrando la frecuencia (o porcentaje) de ocurrencia de cada combinación de categorías en una forma tabular. En Pandas, se utiliza comúnmente para realizar análisis exploratorios y detectar posibles asociaciones o dependencias entre variables.

Por ejemplo, la función `pd.crosstab()` permite generar fácilmente una tabla de contingencia. Esto es muy útil para identificar patrones en los datos, como la distribución de pasajeros por clase y sexo en el dataset del Titanic, entre otros usos.

```python
pd.crosstab(data["gender"], data["education_level"], margins=True)
```

```python
# Tabla de contingencia de Genero, nivel de educación y ciudad
pd.crosstab([data["gender"], data["education_level"]], data["city"], margins=True)
```

```python
# Puntuación promedio por género y nivel de educación

pd.crosstab(
    data["gender"], data["education_level"], values=data["score"], aggfunc="mean"
)
```

## Medidas de asimetría

**Asimetria de Fisher (skewness)**

La asimetría es la medida que indica la simetría de la distribución de una variable respecto a la media aritmética, sin necesidad de hacer la representación gráfica. Los coeficientes de asimetría indican si hay el mismo número de elementos a la izquierda y derecha de la media.

Existen tres tipos de curva de distribución según su asimetría:

* **Asimetría negativa:** la cola de la distribución se alarga para valores inferiores a la media.
* **Simétrica:** hay el mismo número de elementos a izquierda y derecha de la media. En este caso, coinciden la media, la mediana y la moda. La distribución se adapta a la forma de la campana de Gauss, o distribución normal.
* **Asimetría positiva:** la cola de la distribución se alarga para valores superiores a la media.

```python
x.skew()
```

```python
data.skew(numeric_only=True)
```

## Curtosis

Esta medida determina el grado de concentración que presentan los valores en la región central de la distribución. Por medio del Coeficiente de Curtosis, podemos identificar si existe una gran concentración de valores (Leptocúrtica), una concentración normal (Mesocúrtica) ó una baja concentración (Platicúrtica).

```python
data.kurtosis(numeric_only=True)
```

# Transformación de variables

## Columnas Dummy

Convertir variables categóricas a numéricas.

```python
# crear datos categóricos

raw_data = {
    "first_name": ["Jason", "Molly", "Tina", "Jake", "Amy"],
    "last_name": ["Miller", "Jacobson", "Ali", "Milner", "Cooze"],
    "sex": ["male", "female", "male", "female", "female"],
}
df = pd.DataFrame(raw_data, columns=["first_name", "last_name", "sex"])
df
```

```python
# Crear un set de variables dummy para la columna sex

df_sex = pd.get_dummies(df["sex"])
df_sex
```

```python
# Unir los dos dataframes

df_new = df.join(df_sex)

df_new
```

# Leer y guardar datos

Pandas puede leer una variedad de tipos de archivos usando sus métodos `pd.read_`

* csv
* parquet (**formato recomendado**)
* excel
* json
* html
* sql

```python
df = pd.DataFrame({
    "ID": [1, 2, 3, 4, 5],
    "Nombre": ["Ana", "Beto", "Carla", "Daniel", "Elena"],
    "Puntaje": [95, 80, 77, 91, 88]
})

print("DataFrame Original:")
print(df)
```

```python
# 1. Guardar como CSV (archivo de texto plano)
df.to_csv("datos.csv", index=False)

# 2. Guardar como Excel
df.to_excel("datos.xlsx", index=False)

# 3. Guardar como Parquet (formato columnar)
df.to_parquet("datos.parquet", index=False)
```

```python
import os

tamanio_csv = os.path.getsize("datos.csv")
tamanio_excel = os.path.getsize("datos.xlsx")
tamanio_parquet = os.path.getsize("datos.parquet")

print(f"Tamaño CSV    : {tamanio_csv} bytes")
print(f"Tamaño Excel  : {tamanio_excel} bytes")
print(f"Tamaño Parquet: {tamanio_parquet} bytes")
```

* **CSV** suele ser más grande que Parquet, porque no aplica compresión ni guarda metadatos de tipos de datos.

* **Excel** tiende a ser más grande que CSV (con un dataset muy simple puede quedar similar, pero suele aumentar con hojas y formato).

* **Parquet** suele ser mucho más pequeño si los datos se pueden comprimir bien. Además, conserva los tipos de datos de forma nativa y ofrece lectura/escritura eficiente para análisis.
