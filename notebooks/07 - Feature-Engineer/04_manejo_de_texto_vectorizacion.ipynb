# Preprocesamiento de texto: CountVectorizer, TF-IDF y stopwords

Este notebook toma la idea del archivo base y la desarrolla con un ejemplo más largo de clasificación de reseñas.

## **Limpieza**

```python
import re

def clean_text(text):
    # Minúsculas
    text = text.lower()

    # Eliminar URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Eliminar menciones y hashtags
    text = re.sub(r'@\w+|#\w+', '', text)

    # Eliminar puntuación
    text = re.sub(r'[^\w\s]', '', text)

    # Eliminar números
    text = re.sub(r'\d+', '', text)

    # Eliminar espacios múltiples
    text = re.sub(r'\s+', ' ', text).strip()

    return text

text = "Check out https://example.com! @user #AI is awesome!!!"
clean_text(text)
# Resultado: "check out is awesome"
```

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

pd.set_option('display.max_columns', None)
```

## 1. Crear un corpus más amplio

Vamos a simular reseñas de e-commerce con clases positivas y negativas.

```python
positivas = [
    'La entrega fue rápida y el producto llegó en excelente estado',
    'Muy buena calidad, funciona perfecto y lo volvería a comprar',
    'El empaque estaba muy bien, recomendado totalmente',
    'Excelente compra, superó mis expectativas',
    'La atención al cliente fue muy buena y el envío puntual',
    'Producto original, fácil de usar y de buena calidad',
    'Me encantó, se siente premium y cumple lo prometido',
    'Muy satisfecho con la compra, llegó antes de tiempo'
]

negativas = [
    'El producto llegó dañado y la caja abierta',
    'Mala calidad, dejó de funcionar en dos días',
    'El envío fue muy lento y la atención pésima',
    'No cumple con la descripción, muy decepcionante',
    'Se ve barato, materiales malos y mal terminado',
    'La batería dura muy poco, no lo recomiendo',
    'Llegó incompleto y el soporte no respondió',
    'Terrible compra, perdí mi dinero'
]

amplificadores = [
    ' para uso diario', ' en casa', ' en oficina', ' durante el viaje',
    ' y además fue fácil de instalar', ' incluso mejor que en las fotos',
    ' aunque el precio fue alto', ' considerando el precio'
]

rows = []
for i, txt in enumerate(positivas * 25):
    rows.append((txt + amplificadores[i % len(amplificadores)], 1))
for i, txt in enumerate(negativas * 25):
    rows.append((txt + amplificadores[(i + 3) % len(amplificadores)], 0))

df = pd.DataFrame(rows, columns=['texto', 'target'])
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
df.head()
```

## 2. Limpieza básica

```python
stopwords_es = {
    'el','la','los','las','de','del','y','o','en','un','una','unos','unas','muy','con','sin','para',
    'mi','tu','su','lo','al','a','que','fue','fui','por','antes','aunque','ademas','más','mas'
}

def limpiar_texto(texto):
    texto = texto.lower()
    texto = re.sub(r'[^a-záéíóúñü\s]', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

df['texto_limpio'] = df['texto'].apply(limpiar_texto)
df[['texto', 'texto_limpio']].head(10)
```

## 3. Separar train/test

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['texto_limpio'], df['target'], test_size=0.25, random_state=42
)
```

## 4. CountVectorizer sin stopwords

```python
count_vec = CountVectorizer(ngram_range=(1, 2), min_df=2)
Xtr_count = count_vec.fit_transform(X_train)
Xte_count = count_vec.transform(X_test)

print('Shape matriz count:', Xtr_count.shape)
feature_names_count = count_vec.get_feature_names_out()
feature_names_count[:30]
```

```python
counts = np.asarray(Xtr_count.sum(axis=0)).ravel()
top_terms_count = pd.Series(counts, index=feature_names_count).sort_values(ascending=False).head(20)
top_terms_count
```

## 5. CountVectorizer con stopwords

```python
count_vec_sw = CountVectorizer(ngram_range=(1, 2), min_df=2, stop_words=list(stopwords_es))
Xtr_count_sw = count_vec_sw.fit_transform(X_train)
feature_names_count_sw = count_vec_sw.get_feature_names_out()
counts_sw = np.asarray(Xtr_count_sw.sum(axis=0)).ravel()
pd.Series(counts_sw, index=feature_names_count_sw).sort_values(ascending=False).head(20)
```

## 6. TF-IDF

TF-IDF reduce el peso de términos muy frecuentes y resalta términos más distintivos.

```python
tfidf_vec = TfidfVectorizer(ngram_range=(1, 2), min_df=2, stop_words=list(stopwords_es))
Xtr_tfidf = tfidf_vec.fit_transform(X_train)
Xte_tfidf = tfidf_vec.transform(X_test)

print('Shape matriz tfidf:', Xtr_tfidf.shape)
tfidf_features = tfidf_vec.get_feature_names_out()
tfidf_features[:30]
```

## 7. Ver representación sparse

```python
sample_matrix = pd.DataFrame(
    Xtr_tfidf[:5, :15].toarray(),
    columns=tfidf_features[:15]
)
sample_matrix
```

## 8. Entrenar modelos

```python
def train_and_report(Xtr, Xte, ytr, yte, title):
    clf = LogisticRegression(max_iter=3000)
    clf.fit(Xtr, ytr)
    pred = clf.predict(Xte)
    acc = accuracy_score(yte, pred)
    print(f'\n{title}')
    print('Accuracy:', round(acc, 4))
    print(classification_report(yte, pred))
    return clf, acc

clf_count, acc_count = train_and_report(Xtr_count, Xte_count, y_train, y_test, 'Modelo con CountVectorizer')
clf_tfidf, acc_tfidf = train_and_report(Xtr_tfidf, Xte_tfidf, y_train, y_test, 'Modelo con TF-IDF')
```

## 9. Palabras más influyentes en el modelo TF-IDF

```python
coefs = pd.Series(clf_tfidf.coef_[0], index=tfidf_features)
top_pos = coefs.sort_values(ascending=False).head(15)
top_neg = coefs.sort_values().head(15)

print('Más asociadas a clase positiva:')
display(top_pos)
print('Más asociadas a clase negativa:')
display(top_neg)
```

## 10. Comparación final y conclusiones

### Ideas clave

- `CountVectorizer` cuenta ocurrencias.
- `TF-IDF` pondera relevancia relativa.
- Las stopwords ayudan a quitar ruido, aunque no siempre conviene eliminarlas todas.
- Los n-gramas permiten capturar frases, no solo palabras aisladas.
