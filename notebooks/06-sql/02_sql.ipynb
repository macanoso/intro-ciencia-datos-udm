## **Crear tablas en sql**

```python
conn = sqlite3.connect(":memory:")

clientes.to_sql("clientes", conn, index=False, if_exists="replace")
pedidos.to_sql("pedidos", conn, index=False, if_exists="replace")
productos.to_sql("productos", conn, index=False, if_exists="replace")
detalle_pedido.to_sql("detalle_pedido", conn, index=False, if_exists="replace")
```

### **Funciones auxiliares**

```python
def run_sql(query):
    return pd.read_sql_query(query, conn)
```

## **8) Funciones agregadas**


```python
query_8 = """
SELECT 
    COUNT(*) AS numero_pedidos,
    AVG(total) AS promedio_compra,
    MIN(total) AS minimo,
    MAX(total) AS maximo,
    SUM(total) AS suma_total
FROM pedidos
"""
sql_8 = run_sql(query_8)
sql_8
```

```python
pd.DataFrame({
    "numero_pedidos": [pedidos.shape[0]],
    "promedio_compra": [pedidos["total"].mean()],
    "minimo": [pedidos["total"].min()],
    "maximo": [pedidos["total"].max()],
    "suma_total": [pedidos["total"].sum()],
})
```

## **9) GROUP BY**


```python
query_9 = """
SELECT cliente_id, SUM(total) AS total_comprado
FROM pedidos
GROUP BY cliente_id
ORDER BY total_comprado DESC
"""
sql_9 = run_sql(query_9)
sql_9
```

## **10) GROUP BY con HAVING**


```python
query_10 = """
SELECT cliente_id, SUM(total) AS total_comprado
FROM pedidos
GROUP BY cliente_id
HAVING SUM(total) > 200
ORDER BY total_comprado DESC
"""
sql_10 = run_sql(query_10)
sql_10
```

```python
pandas_10 = (
    pedidos.groupby("cliente_id", as_index=False)["total"]
    .sum()
    .rename(columns={"total": "total_comprado"})
)
pandas_10[pandas_10["total_comprado"] > 200].sort_values("total_comprado", ascending=False)
```

## **11) INNER JOIN**

```python
query_11 = """
SELECT 
    c.nombre,
    c.ciudad,
    p.pedido_id,
    p.fecha_pedido,
    p.total
FROM clientes c
INNER JOIN pedidos p
    ON c.cliente_id = p.cliente_id
ORDER BY p.pedido_id
"""
sql_11 = run_sql(query_11)
sql_11.head(10)
```

## **12) LEFT JOIN**

```python
query_12 = """
SELECT 
    c.cliente_id,
    c.nombre,
    p.pedido_id
FROM clientes c
LEFT JOIN pedidos p
    ON c.cliente_id = p.cliente_id
ORDER BY c.cliente_id, p.pedido_id
"""
sql_12 = run_sql(query_12)
sql_12.head(15)
```

```python
(
    clientes.merge(pedidos[["cliente_id", "pedido_id"]], on="cliente_id", how="left")
    .sort_values(["cliente_id", "pedido_id"])
)
```

## **13) JOIN con múltiples tablas**

```python
query_13 = """
SELECT
    c.nombre,
    p.pedido_id,
    pr.nombre_producto,
    dp.cantidad,
    dp.precio_unitario
FROM clientes c
INNER JOIN pedidos p
    ON c.cliente_id = p.cliente_id
INNER JOIN detalle_pedido dp
    ON p.pedido_id = dp.pedido_id
INNER JOIN productos pr
    ON dp.producto_id = pr.producto_id
ORDER BY c.nombre,p.pedido_id
"""
sql_13 = run_sql(query_13)
sql_13.head(15)
```

```python
(
    clientes.merge(pedidos, on="cliente_id", how="inner")
    .merge(detalle_pedido, on="pedido_id", how="inner")
    .merge(productos, on="producto_id", how="inner")
    [["nombre", "pedido_id", "nombre_producto", "cantidad", "precio_unitario"]]
    .sort_values(["nombre","pedido_id"])
)
```

## **14) Ventas por categoría**

```python
query_14 = """
SELECT
    pr.categoria,
    SUM(dp.cantidad * dp.precio_unitario) AS ventas_totales
FROM detalle_pedido dp
INNER JOIN productos pr
    ON dp.producto_id = pr.producto_id
GROUP BY pr.categoria
ORDER BY ventas_totales DESC
"""
sql_14 = run_sql(query_14)
sql_14
```

```python
tmp = detalle_pedido.merge(productos, on="producto_id", how="inner")
tmp["ventas_linea"] = tmp["cantidad"] * tmp["precio_unitario"]

(
    tmp.groupby("categoria", as_index=False)["ventas_linea"]
    .sum()
    .rename(columns={"ventas_linea": "ventas_totales"})
    .sort_values("ventas_totales", ascending=False)
)
```

## **15) Top clientes por monto comprado**

```python
query_15 = """
SELECT
    c.nombre,
    SUM(p.total) AS total_comprado
FROM clientes c
INNER JOIN pedidos p
    ON c.cliente_id = p.cliente_id
GROUP BY c.nombre
ORDER BY total_comprado DESC
LIMIT 5
"""
sql_15 = run_sql(query_15)
sql_15
```

```python
(
    clientes.merge(pedidos, on="cliente_id", how="inner")
    .groupby("nombre", as_index=False)["total"]
    .sum()
    .rename(columns={"total": "total_comprado"})
    .sort_values("total_comprado", ascending=False)
    .head(5)
)
```

## **16) CASE WHEN**

```python
query_16 = """
SELECT
    pedido_id,
    total,
    CASE
        WHEN total >= 200 THEN 'Alto'
        WHEN total >= 100 THEN 'Medio'
        ELSE 'Bajo'
    END AS segmento_compra
FROM pedidos
ORDER BY pedido_id
"""
sql_16 = run_sql(query_16)
sql_16.head(10)
```

```python
import numpy as np

pandas_16 = pedidos[["pedido_id", "total"]].copy()

pandas_16["segmento_compra"] = np.where(
    pandas_16["total"] >= 200, "Alto",
    np.where(
        (pandas_16["total"] >= 100) & (pandas_16["total"] < 200), "Medio", 
        "Bajo"
    )
)

pandas_16.sort_values("pedido_id").head(10)
```

## **17) Subquery**

```python
query_17 = """
SELECT *
FROM pedidos
WHERE total > (
    SELECT AVG(total)
    FROM pedidos
)
ORDER BY total DESC
"""
sql_17 = run_sql(query_17)
sql_17
```

```python
promedio_total = pedidos["total"].mean()
pedidos[pedidos["total"] > promedio_total].sort_values("total", ascending=False)
```

## **18) Query final completa**

```python
query_18 = """
SELECT
    c.nombre,
    c.ciudad,
    pr.categoria,
    SUM(dp.cantidad * dp.precio_unitario) AS gasto_total
FROM clientes c
INNER JOIN pedidos p
    ON c.cliente_id = p.cliente_id
INNER JOIN detalle_pedido dp
    ON p.pedido_id = dp.pedido_id
INNER JOIN productos pr
    ON dp.producto_id = pr.producto_id
GROUP BY c.nombre, c.ciudad, pr.categoria
ORDER BY gasto_total DESC
"""
sql_18 = run_sql(query_18)
sql_18.head(15)
```

```python
tmp = (
    clientes.merge(pedidos, on="cliente_id", how="inner")
    .merge(detalle_pedido, on="pedido_id", how="inner")
    .merge(productos, on="producto_id", how="inner")
)

tmp["gasto_total"] = tmp["cantidad"] * tmp["precio_unitario"]

(
    tmp.groupby(["nombre", "ciudad", "categoria"], as_index=False)["gasto_total"]
    .sum()
    .sort_values("gasto_total", ascending=False)
).head(15)

```
