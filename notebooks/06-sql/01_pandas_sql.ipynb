## **GROUP BY**

```python
#
#  SELECT
#      sex,
#      count(*)
#  FROM
#      df
#  GROUP BY
#      sex;
#
tips.groupby('sex').size()
tips.sex.value_counts()
```

```python
#
#  SELECT
#      day,
#      AVG(tip),
#      COUNT(*)
#  FROM
#      tips
#  GROUP BY
#      day;
#
tips.groupby('day').agg({'tip': np.mean, 'day': np.size})
```

```python
#
#  SELECT
#      smoker,
#      day,
#      COUNT(*),
#      AVG(tip)
#  FROM
#      tips
#  GROUP BY
#      smoker,
#      day;
#
tips.groupby(['smoker', 'day']).agg({'tip': [np.size, np.mean]})
```

## **INNER JOIN**

```python
#
# DataFrames de ejemplo
#
df1 = pd.DataFrame(
    {
        "key_1": ["A", "B", "C", "D"],
        "value": np.random.randn(4)
    }
)

df1
```

```python
df2 = pd.DataFrame(
    {
        'key_2': ['B', 'D', 'D', 'E'],
        'value': np.random.randn(4)
    }
)

df2
```

```python
#
#  SELECT
#      *
#  FROM
#      df1
#  LEFT JOIN
#      df2
#  ON df1.key_1 = df2.key_2;
#
pd.merge(
    df1,
    df2,
    how = "left",
    left_on="key_1",
    right_on = "key_2"
)
```

```python
#
#  SELECT
#      *
#  FROM
#      df1
#  INNER JOIN
#      df2
#  ON df1.key_1 = df2.key_2;
#
pd.merge(
    df1,
    df2,
    left_on="key_1",
    right_on = "key_2"
)
```

## **LEFT OUTER JOIN**

```python
#
#  SELECT
#      *
#  FROM
#      df1
#  LEFT OUTER JOIN
#      df2
#  ON
#      df1.key = df2.key;
#
pd.merge(
    df1,
    df2,
    on="key",
    how="left",
)
```

## **RIGHT OUTER JOIN**

```python
#
#  SELECT
#      *
#  FROM
#      df1
#  RIGHT OUTER JOIN
#      df2
# ON
#     df1.key = df2.key;
#
pd.merge(
    df1,
    df2,
    on="key",
    how="right",
)
```

## **FULL OUTER JOIN**

```python
#
#  SELECT
#      *
#  FROM
#      df1
#  FULL OUTER JOIN
#      df2
#  ON
#      df1.key = df2.key;
#
pd.merge(df1, df2, on='key', how='outer')
```

## **UNION**

```python
df1 = pd.DataFrame(
    {
        'city': ['Chicago', 'San Francisco', 'New York City'],
        'rank': range(1, 4)
    }
)

df1
```

```python
df2 = pd.DataFrame(
    {
        'city': ['Chicago', 'Boston', 'Los Angeles'],
        'rank': [1, 4, 5]
    }
)

df2
```

```python
#
#  SELECT city, rank
#  FROM df1
#  UNION ALL
#  SELECT city, rank
#  FROM df2;
#
pd.concat(
    [df1, df2],
)
```

```python
#
#  SELECT
#      city,
#     rank
#  FROM
#      df1
#  UNION
#      SELECT
#          city,
#          rank
#      FROM
#          df2
#
pd.concat(
    [df1, df2],
).drop_duplicates()
```
